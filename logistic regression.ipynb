```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
data=pd.read_csv("bank full.csv")
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
# One-Hot Encoding of categrical variables
data1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
data1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data1
```

```python
# Custom Binary Encoding of Binary o/p variables 
data1['default'] = np.where(data1['default'].astype(str).str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].astype(str).str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].astype(str).str.contains("yes"), 1, 0)
data1['Target'] = np.where(data1['Target'].astype(str).str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

```python
# Dividing our data into input and output variables
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# As accuracy = 0.8933, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(x)[:,1] 
```

```python
# ROC Curve plotting and finding AUC value
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
